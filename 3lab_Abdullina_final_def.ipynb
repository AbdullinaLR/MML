Решите задачу парной линейной регрессии при помощи модели TensorFlow, рассматривая тензор ранга 1 из пункта 1 задания как значения зависимой переменной (отклика), а точки отрезка из индивидуального задания как значения независимой переменной (предиктора). Предварительно масштабируйте независимую и зависимую переменные на интервал [0, 1]. Оцените качество полученной модели по показателю качества регрессии, указанному в индивидуальном задании ( максимальная ошибка (MaxErr)). Количество эпох, скорость обучения и начальные значения весов выберите самостоятельно, обеспечивая сходимость обучения.  

```python
# Ф-ция для масштабирования переменных на интервал [0, 1]
def scale(x):
    return (x - tf.reduce_min(x)) / (tf.reduce_max(x) - tf.reduce_min(x))

# Масштабирование переменных на интервал [0, 1]
xs = scale(rank_1_tensor) #тензор - завис
ys = scale(func_1) #независ

class Model(object):
    def __init__(self):
        self.w = tf.Variable(tf.constant(tf.random.normal(shape=())))
        self.b = tf.Variable(tf.constant(tf.random.normal(shape=())))

    def __call__(self, x):
        return self.w * x + self.b

model = Model()

# Функция потерь
def loss(predicted_y, target_y):
    return tf.reduce_mean(tf.square(predicted_y - target_y))

def maxerr(y_i, y_hat):
    errors = []
    n = len(y_i)
    for i in range(n):
        error = abs(y_i[i] - y_hat[i])
        errors.append(error)
    max_error = max(errors)
    return max_error

TRUE_w = -0.3
TRUE_b = 0.6

# Функция для обучения модели
def train(model, inputs, outputs, learning_rate):
    with tf.GradientTape() as t:
        current_loss = loss(model(inputs), outputs)
    dw, db = t.gradient(current_loss, [model.w, model.b])
    model.w.assign_sub(learning_rate * dw)
    model.b.assign_sub(learning_rate * db)

    return current_loss

list_w, list_b = [], []
epochs = range(151)
losses = []
max_errors = []
for epoch in epochs:
    list_w.append(model.w.numpy())
    list_b.append(model.b.numpy())
    current_loss = train(model, xs, ys, learning_rate=0.1)
    losses.append(current_loss)
    predicted_xs = model(xs)
    max_error = maxerr(ys, predicted_xs).numpy()
    max_errors.append(max_error)  # Сохранение максимальной ошибки после каждой эпохи
    if epoch % 10 == 0:
        print('Эпоха %4d: w=%1.2f b=%1.2f, потери=%2.5f' %
              (epoch, list_w[-1], list_b[-1], current_loss))
        print('Макс ошибка данной эпохи:', max_error)


print('Максимальная ошибка (MaxErr):', max_error)
print('Текущие потери (ошибка): %1.6f' % loss(model(xs), ys).numpy())
```

```python
# Визуализация обучения
plt.plot(epochs, list_w, 'r', epochs, list_b, 'b')
plt.plot([TRUE_w] * len(epochs), 'r--', [TRUE_b] * len(epochs), 'b--')
plt.legend(['w', 'b', 'Истинное w', 'Истинное b'])
plt.show()
```
